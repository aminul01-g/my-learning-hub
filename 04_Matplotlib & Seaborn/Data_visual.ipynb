#### **âœ… Step 1: Setup**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic.csv')
```

#### **ðŸ”¹ 1. Histogram â€” Age Distribution**


```python
plt.hist(df['Age'].dropna(), bins = 20, edgecolor = 'black')
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

#### **ðŸ”¹ 2. Bar Plot â€” Survival Count**

```python
df['Survived'].value_counts().plot(kind = 'bar')
plt.title("Survived Count")
plt.xticks([0,1],['Not Survived','Survived'])
plt.ylabel("Number of passaengers")
plt.show()
```

#### **ðŸ”¹ 3. Seaborn: Countplot â€” Survival by Gender**

```python
sns.countplot(data = df, x = 'Sex', hue = 'Survived')
plt.title("Survival by Gender")
plt.legend(title = 'Survived', labels = ['NO', 'Yes'])
plt.show()
```

#### **ðŸ”¹ 4. Seaborn: Boxplot â€” Age by Passenger Class**

```python
sns.boxplot(data = df, x = 'Pclass', y = 'Age')
sns.title("Age distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

#### **ðŸ”¹ 5. Heatmap â€” Correlation Matrix**

df.corr():
 Computes the correlation matrix of the DataFrame.
Correlation measures how strongly two numeric variables are related (ranges from -1 to 1).

* 1 means perfect positive correlation (they increase    together).

* -1 means perfect negative correlation (one increases while the other decreases).

* 0 means no correlation.

sns.heatmap(corr, annot=True, cmap='coolwarm')

* sns.heatmap(...): Plots the correlation matrix as a heatmap â€” a colored grid.

* corr: The matrix of correlations you just computed.

* annot=True: Writes the actual correlation values (numbers) inside the boxes.

* cmap='coolwarm': A color scheme where:

* Red shades = negative correlation

* Blue shades = positive correlation

* White = around zero

```python
# Computes the correlation matrix of the DataFrame.
corr = df .corr(numeric_only = True)
sns.heatmap(corr,  annot = True, cmap = 'coolwarm')
plt.title("Correlation matrix")
plt.show()
```
