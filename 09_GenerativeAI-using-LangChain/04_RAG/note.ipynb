# Problems in RAG

#### Quick Recap  What RAG Does

RAG (Retrieval-Augmented Generation) = Search + Generate

It helps the LLM answer by:

1. Searching external data (your files, DB, etc.)

2. Feeding the retrieved info + user question into the model

So the model doesnt need to know the info inside  it just reads it each time.

 Advantages:

- No need to retrain the model

- Easy to update data

- Reduces hallucinations
